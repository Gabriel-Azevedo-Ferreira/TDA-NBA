```python
import pandas as pd
import numpy as np
df = pd.read_csv('../../Data/NBA_totals_with_positions.csv')
print "(rows,cols) =",df.shape
df.head(5)
```

```python
dict = {' PG': 0, ' PF': 1, ' SG' : 2, ' SF' : 3, ' C' : 4}

position_labels = [dict[x] for x in df['PosicAbbrev']]
df_numVar = df.drop(['Unnamed: 0', 'PLAYER', 'TEAM'],1)
df_numVar.head()
```

```python
df_numVar.columns
```

```python
df_numVar.iloc[:,5:26] = df_numVar.iloc[:,5:26].div(df_numVar["MIN"], axis=0) #dividing by minutes
df_numVar[["FG%","3P%","FT%"]] = df_numVar[["FG%","3P%","FT%"]].multiply(df_numVar["MIN"], axis = 0) #not dividing the % ones
```

```python
posit_numbers = df_numVar['PosicAbbrev'].values
df_numVar['PosicNumber'] = [dict[x] for x in posit_numbers]
df_numVar = df_numVar.drop('PosicAbbrev', 1)
```
